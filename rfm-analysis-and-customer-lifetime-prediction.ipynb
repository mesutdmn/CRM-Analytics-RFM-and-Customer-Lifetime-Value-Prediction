# <p style="background-color:#F8F1E8; font-family:newtimeroman;color:#602F44; font-size:150%; text-align:center; border-radius: 15px 50px;"> üßê RFM Analysis üîç</p>

```python
rfm = data[["ID","Recency","frequency","monetary"]]
rfm.head()
```

<div style="border-radius:10px; border:#602F44 solid; padding: 15px; background-color: #F6C9BC; font-size:100%; text-align:left">

<h3 align="left"><font color='#602F44'>üìù Notes:</font></h3>
Created RFM DataFrame based on "ID","Recency","frequency","monetary"

```python
rfm["recency_score"] = pd.qcut(rfm["Recency"], q=5 , labels=[5,4,3,2,1])
rfm["frequency_score"] = pd.qcut(rfm["frequency"], q=5 , labels=[1,2,3,4,5])
rfm["monetary_score"] = pd.qcut(rfm["monetary"], q=5 , labels=[1,2,3,4,5])
rfm.head()
```

<div style="border-radius:10px; border:#602F44 solid; padding: 15px; background-color: #F6C9BC; font-size:100%; text-align:left">

<h3 align="left"><font color='#602F44'>üìù Notes:</font></h3>
Created RFM Scores and named them recency_score	frequency_score	monetary_score

```python
rfm["RF_SCORE"] = (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
regex_dict = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_Risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
}
rfm['segment'] = rfm['RF_SCORE'].replace(regex_dict, regex=True)
```

<h3 align="left"><font color='#602F44'>üìù Notes:</font></h3>
Created regex dictionary and used it for segment our customers based on their RF score

```python
import datetime as dt
data["last_order_date"] = pd.to_datetime(-data["Recency"], unit="D", origin=dt.datetime(2014,10, 4))
```

<div style="border-radius:10px; border:#602F44 solid; padding: 15px; background-color: #F6C9BC; font-size:100%; text-align:left">

<h3 align="left"><font color='#602F44'>üìù Notes:</font></h3>
Created a last order date feature from analyis day - recency

```python
cltv = pd.DataFrame()
cltv["Recency_weekly"] = (data["last_order_date"] - data["Dt_Customer"]).dt.days // 7
cltv["T_weekly"] = (dt.datetime(2014,10, 4) - data["Dt_Customer"]).dt.days // 7
cltv["frequency"] = data["frequency"]
cltv["monetary_avg"] = data["monetary"] // data["frequency"]
cltv.index = data["ID"]
```

<div style="border-radius:10px; border:#602F44 solid; padding: 15px; background-color: #F6C9BC; font-size:100%; text-align:left">

<h3 align="left"><font color='#602F44'>üìù Notes:</font></h3>
Created recency_weekly, t_weekly, frequency, monetary_avg for our BD/NBD and Gamma Gamma Modeling

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(y="monetary_avg", data=cltv,palette="pastel");
plt.subplot(2,2,2)
sns.boxplot(y="frequency", data=cltv,palette="pastel");
```

```python
def outliner_fixer(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    up = q3 + 1.5 * IQR
    low = q1 - 1.5 * IQR
    return round(up),round(low)
def limit_it(df,col):
    up,low = outliner_fixer(df,col)
    df.loc[df[col] > up, col] = up
    df.loc[df[col] < low, col] = low
```

```python
limit_it(cltv,"frequency")
limit_it(cltv,"monetary_avg")
```

### <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#8581B8; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid">BD/NBD Modeling</p>

```python
!pip install lifetimes
from lifetimes import BetaGeoFitter
from lifetimes import GammaGammaFitter
from lifetimes.plotting import plot_period_transactions
```

```python
bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(cltv['frequency'],
        cltv['Recency_weekly'],
        cltv['T_weekly'])
```

```python
#predictions for next 3 months and then next 6 months
cltv["exp_sales_3_month"] = bgf.predict(4 * 3,
                                        cltv['frequency'],
                                        cltv['Recency_weekly'],
                                        cltv['T_weekly'])
cltv["exp_sales_6_month"] = bgf.predict(4 * 6,
                                        cltv['frequency'],
                                        cltv['Recency_weekly'],
                                        cltv['T_weekly'])
```

```python
cltv["exp_sales_3_month"].sort_values(ascending=False).head(10)
```

### <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#8581B8; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid">Gamma Gamma Modeling</p>

```python
ggf = GammaGammaFitter(penalizer_coef=0.01)
ggf.fit(cltv['frequency'],
        cltv['monetary_avg'])
```

```python
cltv["exp_average_value"] = ggf.conditional_expected_average_profit(cltv['frequency'],
                                                                    cltv['monetary_avg'])
```

```python
cltv["exp_average_value"].sort_values(ascending=False).head(10)
```

### <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#8581B8; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid">Customer Lifetime Value</p>

```python
cltv["clv"] = ggf.customer_lifetime_value(bgf,
                                   cltv['frequency'],
                                   cltv['Recency_weekly'],
                                   cltv['T_weekly'],
                                   cltv['monetary_avg'],
                                   time= 6,  # months
                                   freq="W" #depends on your data, I set it weekly
                                         )
```

```python
cltv["clv"].sort_values(ascending=False).head(10)
```

```python
cltv["cltv_segment"] = pd.qcut(cltv["clv"], 4, labels=["D", "C", "B", "A"])
```

<h3 align="left"><font color='#DEB887'>üìù Notes:</font></h3>

1. **Segment A (High-Value Customers):**
   - **Description:** Customers in this segment are considered high-value, meaning they contribute significantly to the company's revenue and profitability.
   - **Characteristics:** Typically, these customers make frequent purchases, spend more per transaction, and exhibit strong loyalty.

2. **Segment B (Medium-Value Customers):**
   - **Description:** Customers in this segment have a moderate impact on the overall revenue and profitability of the business.
   - **Characteristics:** They may not be as high-spending or frequent as Segment A customers, but they still contribute positively to the company's bottom line.

3. **Segment C (Low-Value Customers):**
   - **Description:** Customers in this segment have a lower impact on revenue compared to A and B segments.
   - **Characteristics:** They may make infrequent purchases or have lower transaction values. The goal may be to engage them more effectively to increase their value over time.

4. **Segment D (Churn or At-Risk Customers):**
   - **Description:** Customers in this segment are at risk of churning or have already churned.
   - **Characteristics:** They may exhibit a decline in purchasing frequency or spend, and the company needs to implement retention strategies to prevent them from leaving.
